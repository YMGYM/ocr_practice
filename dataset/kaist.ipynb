# Kaist 데이터셋 편집

KAIST 데이터셋을 학습 가능한 형태로 편집하는 코드입니다.

```python
import os, sys
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from xml.etree.ElementTree import parse
```

```python
data_dir = 'files/raw/'
```

```python
all_data = os.listdir(data_dir)
```

```python
xml_data = [x for x in all_data if (x[-3:] == 'xml' or x[-3:] == 'XML')]
```

```python
len(xml_data) # 더 적은쯕으로 사용
```

# XML 파일 오픈
데이터셋 설정을 위해 XML 파일을 열어야 합니다.

```python
tree = parse(data_dir + xml_data[0])
```

```python
imgName = tree.find("imageName")
```

```python
imgName = tree.getroot().find("./image/imageName").text
```

```python
img = Image.open(data_dir + imgName)
plt.imshow(img)
plt.show()
```

```python
words = tree.getroot().find("./image/words")
```

```python
for w in words:
    print(w.find('./character').text)
```
